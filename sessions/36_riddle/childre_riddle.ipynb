```julia
possibilities = [] 
```

```julia
for bl in 1:2 
    for gr in (bl+1):5
        for br in (gr+1):7 
            for sm in (br+1):11
                if sum([bl, gr, br, sm]) <= 17
                    push!(possibilities, (bl, gr, br, sm))
                end
            end
        end
    end
end
```

```julia
possibilities
```

```julia
house_nr = [prod(x) for x in possibilities]

```

```julia
unique(house_nr)
```

```julia
candidates =[(i, count(==(i), house_nr)) for i in unique(house_nr) if count(==(i), house_nr) > 1]
```

```julia
result = []
for (candidate, cnt) in candidates
    push!(result, [])
    for combo in possibilities
        if prod(combo) == candidate
            push!(result[end], (candidate, combo))
        end
    end
end
```
