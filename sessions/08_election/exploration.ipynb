```r
library(tidyverse)
```

```r
data <- read_tsv("kerg2.csv", skip = 8, locale=locale(decimal_mark = ","))
```

```r
glimpse(data)
```

```r
data %>% filter(str_detect(Gebietsname, "Würz"))
```

```r
data %>%
    filter(str_detect(Gruppenname, "GRÜNE"), Stimme==2, Gebietsart=="Wahlkreis") %>%
    ggplot(aes(VorpProzent, Prozent)) +
        geom_point() +
        geom_abline(slope=1, intercept=0)
```

```r
data %>%
    filter(str_detect(Gruppenname, "GRÜNE"), Stimme==2, Gebietsart=="Wahlkreis", VorpProzent>Prozent)
```
