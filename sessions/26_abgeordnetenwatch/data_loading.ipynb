```r
library(jsonlite)
```

```r
js <- fromJSON("polls.json")
```

```r
tb <- js$data %>% as_tibble(.name_repair = "minimal")
```

```r
tb %>% print
```

```r
load_vote <- function(id){
    result <- fromJSON(str_c("votes/",id,".json"))$data %>% as_tibble(names_repair="minimal")
    result <- result %>% unnest_wider(c("mandate","poll","fraction"), names_sep="_") %>% select(id, poll_id, mandate_label, poll_label, vote, reason_no_show, fraction_label)
    return(result)
}
```

```r
load_vote(4293)
```

```r
all_votes <- tb$id %>% map_df(load_vote)
```

```r
all_votes <- all_votes %>% left_join(select(tb, poll_id=id, poll_date=field_poll_date), by="poll_id")
```
